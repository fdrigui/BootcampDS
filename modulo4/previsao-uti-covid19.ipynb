```python
import numpy as np # linear algebra
from sklearn.linear_model import LogisticRegression
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
dados = pd.read_excel(r'/kaggle/input/covid19/Kaggle_Sirio_Libanes_ICU_Prediction.xlsx')
dados.head()
```

```python
# O modelo de regressão logística não trabalha com variáveis do tipo string. Por isso vamos fazer uma "gambiarra" para eliminar esses valores
X_columns = dados.describe().columns

# O modelo de regressão logística não trabalha com variáveis do tipo NaN ou infinitos, por isso precisamos removelos.
# Nós precisamos que X e y tenham a mesma quantidade de linhas, por isso nós vamos fazer esse "drop" antes de dividir X e y de 'dados'
dados_limpos = dados.dropna()
```

```python
# Modelol é uma função, e como toda função, normalmente dizemos que f(x) = y
# Por esse motivo, as variáveis de entrada são representadas pela letra x e as de saída pela letra y
# Normalmente, quando se tem uma única coluna como entrada, utiliza-se o x minúsculo e maiúsculo quando se tem multiplas colunas
# O mesmo acontece com a letra y;
# Vamos utilizar o X (Maiúsxulo) para entrada, uma vez que temos multiplas colunas de entrada e y (minúsculo) para saída, uma vez que ela é única
y = dados_limpos['ICU']
X = dados_limpos[X_columns].drop(['ICU'], axis=1)
```

```python
modelo = LogisticRegression()
modelo.fit(X, y)
```
